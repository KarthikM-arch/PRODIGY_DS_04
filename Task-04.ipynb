```python
import pandas as pd

# Load the data
data = pd.read_csv('twitter_training.csv')
```

```python
# Clean the data by removing irrelevant columns and duplicates
data_cleaned = data.iloc[:, 1:4]  # Keep columns: Topic, Sentiment, and Text
data_cleaned.drop_duplicates(inplace=True)

# Rename columns for clarity
data_cleaned.columns = ['Topic', 'Sentiment', 'Text']
```

```python

# Display the first few rows of the cleaned data
data_cleaned.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up visual style
sns.set(style="whitegrid")

# Sentiment distribution across all data
plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment', hue='Sentiment', data=data_cleaned, palette='coolwarm', legend=False)
plt.title('Overall Sentiment Distribution', fontsize=14)
plt.xlabel('Sentiment', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

```python
# Sentiment distribution by Topic
plt.figure(figsize=(12, 8))
sns.countplot(x='Topic', hue='Sentiment', data=data_cleaned, palette='coolwarm')
plt.title('Sentiment Distribution by Topic', fontsize=14)
plt.xlabel('Topic', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
